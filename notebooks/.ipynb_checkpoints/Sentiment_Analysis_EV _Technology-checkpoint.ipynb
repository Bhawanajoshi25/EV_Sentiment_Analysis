# Preprocessing The Dataset

```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from langdetect import detect
import emoji
```

```python
df = pd.read_csv('../data/reddit_ev_comments.csv')
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # Remove URLs, mentions, and hashtags
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)  # Removes URLs
    text = re.sub(r'@\w+', '', text)  # Removes mentions
    text = re.sub(r'#\w+', '', text)  # Removes hashtags
    
    # Remove special characters and punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Convert to lowercase
    text = text.lower()  # Ensures all text is lowercase
    
    # Remove extra whitespaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove stopwords
    text = ' '.join(word for word in text.split() if word not in stop_words)
    
    # Lemmatize words
    text = ' '.join(lemmatizer.lemmatize(word) for word in text.split())
    
    return text


def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False

def remove_emojis(text):
    return ''.join(char for char in text if char not in emoji.EMOJI_DATA)
```

```python
df = df[df['Comment_Body'].apply(is_english)]
```

```python
df['Processed_Comment_Body'] = df['Comment_Body'].apply(preprocess_text)
df['Processed_Comment_Body'] = df['Processed_Comment_Body'].apply(remove_emojis)
```

```python
df.to_csv('../data/reddit_ev_preprocessed.csv', index=False)
```

# Sentiment Analysis using Textblob

```python
import pandas as pd
```

```python
!pip install textblob
```

```python
from textblob import TextBlob

# Apply sentiment analysis to each comment using TextBlob
df['Sentiment_Score'] = df['Processed_Comment_Body'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Classify sentiment based on the polarity score
def classify_sentiment(score):
    if score > 0:
        return 'Positive'
    elif score < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['Sentiment_Score'].apply(classify_sentiment)

# Display the results
print(df[['Processed_Comment_Body', 'Sentiment']])
```

```python
df.to_csv('../data/sentiment_analysis_result.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Count the number of comments in each sentiment category
sentiment_counts = df['Sentiment'].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(8, 6))
sentiment_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.xlabel('Sentiment')
plt.ylabel('Number of Comments')
plt.title('Sentiment Distribution of Reddit Comments')
plt.xticks(rotation=0)
plt.show()
```

# Topic Modeling Using LDA (Latent Dirichlet Allocation)

```python
negative_comments_df = df[df['Sentiment'] == 'Negative']
```

```python
!pip install gensim nltk
```

```python
import gensim
from gensim import corpora
from nltk.corpus import stopwords
import nltk
```

```python
# Tokenize comments
negative_comments_df['Tokens'] = negative_comments_df['Processed_Comment_Body'].apply(lambda x: [word for word in x.split() if word not in stop_words])
```

```python
# Create a dictionary from the tokenized words
dictionary = corpora.Dictionary(negative_comments_df['Tokens'])

# Filter out extremes to remove very common and very rare words
dictionary.filter_extremes(no_below=5, no_above=0.5)

# Convert tokenized words to a bag-of-words format
corpus = [dictionary.doc2bow(tokens) for tokens in negative_comments_df['Tokens']]
```

```python
from gensim.models.ldamodel import LdaModel

# Number of topics to extract
NUM_TOPICS = 5

# Train LDA model
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=NUM_TOPICS, random_state=42, passes=10)

# Print the topics discovered by LDA
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx + 1}: {topic}")
```
